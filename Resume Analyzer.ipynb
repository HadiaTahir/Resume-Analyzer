```python
import re
```

```python
def CleanResume(txt):
    # Remove URLs
    CleanTxt = re.sub(r'http\S+', ' ', txt)
    # Remove Retweets, and other shorthand notations like 'RT' and 'CC'
    CleanTxt = re.sub(r'\bRT\b|\bCC\b', ' ', CleanTxt)
    # Remove mentions (e.g., @username)
    CleanTxt = re.sub(r'@\S+', ' ', CleanTxt)
    # Remove hashtags (e.g., #hashtag)
    CleanTxt = re.sub(r'#\S+', ' ', CleanTxt)
    # Remove non-ASCII characters
    CleanTxt = re.sub(r'[^\x00-\x7F]', ' ', CleanTxt)
    # Remove punctuation
    CleanTxt = re.sub(r'[!"#$%&\'()*+,\-./:;<=>?@[\\]^_`{|}~]', ' ', CleanTxt)
    # Replace multiple spaces with a single space
    CleanTxt = re.sub(r'\s+', ' ', CleanTxt).strip()
    return CleanTxt
```

```python
df['Resume'] = df['Resume'].apply(lambda x: CleanResume(x))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words = 'english')
```

```python
tfidf.fit(df['Resume'])
RequiredText = tfidf.transform(df['Resume'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(RequiredText, df['Category'], test_size = 0.2, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf =OneVsRestClassifier( KNeighborsClassifier())
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```
